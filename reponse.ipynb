```python
import re

def remove_references(phrase):
    # Regular expression pattern to match references
    pattern = r"\[.*?\]|\(.*?\)"
    
    # Remove references using regex substitution
    clean_phrase = re.sub(pattern, "", phrase)

    return clean_phrase
```
